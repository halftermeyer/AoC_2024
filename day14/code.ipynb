```python
import re
```

```python

def gen_robots(file='input.txt', part=1):
    """Generates tuples of integers"""
    file = open(file, 'r')

    for ix, line in enumerate(file):
        group =re.match(r'^p=(-?\d+),(-?\d+) v=(-?\d+),(-?\d+)$', line)
        p, v= (int(group[1]), int(group[2])), (int(group[3]), int(group[4]))
        size = (11,7) if part == 1 else (101,103) 
        yield p, v, size
```

```python
filename, part = "input.txt", 2
#filename, part = "test.txt", 1
```

```python
from collections import defaultdict
quad_num = defaultdict(int)
secs = 100
for p, v, size in gen_robots(filename, part):
    print(p, v, size)
    final_pos = tuple((x + u * secs)%s for x, u, s in zip(p, v, size))
    print(final_pos)
    q = tuple((f<((s+1)/2-1), f==((s+1)/2-1), f>((s+1)/2-1)) for f, s in zip(final_pos, size))
    quad_num[q] += 1
part1 = 1
for k in quad_num:
    print(k, quad_num[k])
    if not (k[0][1] or k[1][1]):
        print("ok")
        part1 *= quad_num[k]

part1
```

```python
maze = [[0]*size[0] for _ in range(size[1])]
def show_maze(maze):
    for row in maze:
        print(''.join([' ' if x==0 else '#' for x in row]))
show_maze(maze)
```
